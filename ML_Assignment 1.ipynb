```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/user5/Downloads/used_cars_data (1).csv")

df['Brand'] = df['Name'].apply(lambda x: x.split()[0])

brand_counts = df['Brand'].value_counts()

plt.pie(brand_counts, labels=brand_counts.index, startangle=140)
plt.title('Car Distribution by Brand')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/user5/Downloads/used_cars_data (1).csv")

top_cars = data[['Name', 'Kilometers_Driven']].sort_values(by='Kilometers_Driven', ascending=False).head(10)

plt.bar(top_cars['Name'], top_cars['Kilometers_Driven'], color='mediumseagreen')
plt.xlabel('Car Name')
plt.ylabel('Kilometers Driven')
plt.title('Top 10 Cars by Kilometers Driven')

plt.savefig('C:/Users/user5/Downloads/bar.jpg')
plt.show()
```
